# Numeral systems

## Def
Numeral system is a system of symbolic representation or encoding of numbers.

## History
The first numeral system introduced by Egiptions is a [unary numeral system](https://en.wikipedia.org/wiki/Unary_numeral_system), in contrast of modern most used num. sys. it does not rely on symbol position. In unary num. sys. each natural number was represented by a coressponding number of symbols. For example if the symbol / is chosen, then the number seven would be represented by ///////.

In order of making count related problems easier, the first positional num. sys. was created. It was a 60 based num. sys. created by Babylonians, 60 minutes in an hour and 60 seconds in a minute is nothing else than the babylonian legacy used until these days.

>In positional number system the value of a digit is depending of it's position. And each position represents a power of the base and each digit tells how many of those units you have

## Bases

Base or radix (root from Latin) is a number of unique digits and/or symbols used to represent a number in distinct num. sys.

Most used are radix 10, 2, 16 and 8. Their usability is defined by practical usage. For example the num. sys. of the base 2 is used in the field of electronics as the computer transistor has only two positions: on of off. Base 10 is very practical to count with human fingers.    

## Conversions

### Memo

$A_{10} => B_x$ - divide A by the base of B, save the rest, repeat with result until 0, write rests in one line form right to left  
$A_x => B_{10}$ - multiply each digit of A by $10^{position}$ and sum everything or use the horner method  
$A_x => B_y$ - bridge A to base 10 and use the method to convert base 10 to base y  
$A_x => B_y$ - x and y are powers of 2, use shortcut of grouping

#### Horner's method

A method of decoding a number by it's decomposing.

Let's take $1234_5$, to convert it to decimal base we can take ask ourselves a question, what does $1_5$ will be?

$1_5 => 1_{10}$

And what will be $12_5$ ?

$12_5 => 10_5 + 2_5 => 1 \times 5_{10} + 2_{10} => 7_{10}$  

...

$123_5 => 120_5 + 3 => 12_5 \times 10_5 + 3 = 7 \times 5 + 2 => 38_{10}$  
$1234_5 => 1230 + 4 => 123_5 \times 10_5 + 4 => 38 \times 5 + 3 => 194_{10}$

## Calculations

Any calculations has the same basic principles as in the base of 10, but you should rememember that the overflowing of the place value will depend of the numeral system. In the base 10 the overflowing of the place value will be after 9, in base 2 after 1, in base 6 after 5 and etc. 

### Addition and substraction

For addition of the base of $N$ *if the sum >= $N$, we write $result - N$ and carry 1 to the next column*  
For example, $456_{10} + 789_{10}:$

Column by column:
- $6 + 9 = 15$, 15 > $N$ (10 in that case) so we write $15 - N$ and carry 1 to the next column  
- $5 + 8 = 13$, again bigger, we write 3 + 1 we carried and carry 1 to the next col
- $4 + 7 = 11$, write 1 + 1 we carried, and carry 1 again
In sum: 1245

For substraction the inverted rules are applied, so *if the difference is less than the smallest value (0 in radix 10, 2, 6 etc.), we write $result + N$ we borrow 1 of the next column*  
For example, $1245_{10} - 789_{10}:$

Column by column:
- $5 - 9 = -4$, -4 < $0$ so we write 10 + (-4) => 6 and borrow 1 of the next column  
- $(4 - 1) - 8 = -5$, 10 + (-5) => 5 and borrow 1 of the next column  
- $(2 - 1) - 7 = -6$, write 4 and borrow 1  
In sum: 456

**For any numeral system these rules are working the same**, but for the simplicity we can convert in between to the base of 10 and back

$4532_6 + 123_6$

Column by column:
- $2 + 3 = 5_{10}$, keep 5
- $3 + 2 = 5_{10}$, keep 5
- $5 + 1 = 6_{10}$, as 6 >= $N$ write 0 and carry 1

In sum: $5055_6$
