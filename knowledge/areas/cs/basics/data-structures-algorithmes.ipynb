## Arrays

Array is a collection of elements. Each element can be queried by it's index.

### Copy elements into another array

In programming languages there are another built in solutions, but here is the simpliest implementation

```python
A = [1, 2, 3, 4, 5]
B = []

for k in range(len(A)):
    B[k] = A[k]
```

### Linear array search

```python
def array_search(arr, x):
    """
    Accept an array and value to find. Return index of a first found value in array, or return -1.
    """
    for k in range(len(arr)):
        if arr[k] == x:
            return k
        
    return -1
```

### Array inversion

Move elements of the right half to the left half and vise versa.

```python
def invert_array(arr):
    for k in range(arr // 2):
        arr[k], arr[len(arr) - 1 - k] = arr[len(arr) - 1 - k], arr[k]
```

### Circular shift

Move elements one cell right or one cell left, so the current element takes place of the next one or of the previous one.

```python
# more efficient using shift
def shift_arr_left_shift(arr):
    first_el = arr[0]

    for i in range(len(arr) - 1):
        arr[i] = arr[i + 1]

    arr[len(arr) - 1] = first_el

    return arr

def shift_arr_right_shift(arr):
    last_el = arr[len(arr) - 1]

    for i in range(len(arr) - 2, -1, -1):
        arr[i + 1] = arr[i]

    arr[0] = last_el

    return arr

# less efficient using pair swap
def shift_arr_left_swap(arr):
    for i in range(len(arr) - 1):
        arr[i + 1], arr[i] = arr[i], arr[i+1]

    return arr

def shift_arr_right_swap(arr):
    for i in range(len(arr) - 1, 0, -1):
        arr[i - 1], arr[i] = arr[i], arr[i - 1]

    return arr
```

### Sieve of Eratosthenes

Algorithme for finding primal numbers up to any given number. Any prime number can be divided by 1 or by itself, the first prime number is 2.  
The algorithme is searching for the primal number, throw all next numbers divided by that number by simply selecting n * 2, n * 3, n * 4 etc., and repeat the process.    

```python
def sieve(n):
    A = [True] * n
    A[0] = A[1] = False # as 0 and 1 are not counted as primal numbers

    for k in range(2, n):
        if A[k]:
            for m in range(k*2, n, k):
                A[m] = False
```
