# Python syntax

## Basic        

```python
# variables
x = 2
a = "Hello world"

# function
print(a)

# vars exchange
a = 2
b = 5

a,b = b,a

print(a, b) # 5 2

# cascade variables assignement
x,y,z = 1,2,3 # x -> 1; y -> 2; z -> 3

# multiple variables assignement
x = y = z = 0 # x -> 0; y -> 0; z -> 0
```

## Math

```python
# power
x = 2 ** 3 # 8

# multiplication
x = 2 * 2 # 4

x + 2
x - 2   

# syntax sugar
x -= 1 # x = x - 1
x += 1 # x = x + 1

# division
x / 2 # regular divison
x % 2 # rest of divison
x // 2 # diviosn without a rest 
```

**Important**: in python as in math (but not as in all other languages) the divison rest of a negative number is calculated from the *lower* number.  

` print(-12 // 5 == - 3) # True` *as the lower number which can be divided by 5 without a rest is -15*


```python
print(-12//5)
```

## Advanced math

```python
# calc square root
x = 2
print(x ** 0.5)

# powers
# is calculated from right to left
print(2**3**4)
```

## Operations

Iteration is a single execution

```python
# while
while x > 0:
    print('x is greater than 0')
else:
    print('x is lower than 0')

# if 
if True:
    print(True)
else:
    print(False)

if x > 0:
    print('x is greater than 0')
elif x > -5:
    print('x is greater than -5')

if x < 0 and x > -5:
    print('x is between 0 and -5')
    
# for
for x in 1,2,3:
    print(x)
```

## Functions

```python
def myFunc():
    print('yo')

def myFunc2(num1, num2):
    return num1 + num2

myFunc()
myFunc2(2,2)

# parameters
def greetings(name = "World", separator = " "):
    print("Hello", name, sep = separator)

greetings("Andrei")
greetings(separator="-")
```

## Built-in useful functions 

```python
# range(start, stop, step)
# create a number sequence
for x in range(1,10,1):
    print(x) 

for y in range(10,1,-1):
    print(y)
```
